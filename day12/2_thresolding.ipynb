<h1>CV2</h1>

```python
#by using cv2 funtion
img = cv2.imread('../datasets/images/lenna.png',0)
thes,new_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #thresold image using cv2
new_img,thes
```

```python
#by using array operation
img_bin = (img>127).astype(np.uint) * 255 
img_bin
```

```python
#inverse binary operation
thres_inv,inv_binary_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
inv_binary_img
```

```python
#plotting all

plt.figure(figsize = (16,9))

plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(img,cmap='gray')

plt.subplot(1,3,2)
plt.title('By Array operation')
plt.imshow(img_bin,cmap='gray')

plt.subplot(1,3,3)
plt.title('By CV2 function')
plt.imshow(new_img,cmap='gray')
```
