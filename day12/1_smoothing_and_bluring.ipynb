<h1>Averaging Kernel</h1>

```python
img = cv2.imread('../datasets/images/dog.jpg')
plt.imshow(img)
img.shape
```

```python
#kernel 
kernel = np.zeros([3,3]) + 1/9
kernel
```

```python
blured_img = cv2.filter2D(img,-1,kernel)
cv2.imshow('blur image',blured_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.zeros([9,9]) + 1/81
blured_image = cv2.filter2D(img,-1,kernel)
blured_image = np.hstack([img,blured_image])
```

```python
cv2.imshow('9*9 matrix',blured_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h1>Gaussian Kernel</h1>

```python
kernel_size = 7
kernel = cv2.getGaussianKernel(kernel_size,1.23)
kernel
```

```python
gaussian_blurred_image = cv2.filter2D(img,-1,kernel)
gaussian_blurred_image = np.hstack([img,gaussian_blurred_image])
```

```python
cv2.imshow('gaussian image',gaussian_blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h1>Median Kernel</h1>

```python
img_gray = cv2.imread('../datasets/images/dog.jpg')
```

```python
median_img = cv2.medianBlur(img_gray,15)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap = 'gray')

plt.subplot(1,2,2)
plt.imshow(median_img,cmap = 'gray')
```

<h1>Blurring specific part</h1>

```python
img = plt.imread('../datasets/images/dog.jpg',0)

avg_kernel = np.zeros([9,9]) + 1/81
part = img[100:200,150:250]
blured_part = cv2.filter2D(part,-1,avg_kernel)
img_copy = img.copy()
img_copy[100:200,150:250] = blured_part

plt.imshow(img_copy)
```
