```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../datasets/images/circle.png',0)
img
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img,blockSize = 2,ksize =3 , k = 0.04)
```

```python
corners.min()
```

```python
plt.imshow(corners,cmap = 'gray')
```

```python
img_new = img.copy()
img_new[corners > 0.01 * corners.max()] = [255]
```

```python
plt.imshow(img_new,cmap='gray')
img_new = corners < corners.max() / 2
plt.imshow(img_new,cmap = 'gray')
```

<H1>Lenna image</H1>

```python
lenna = cv2.imread('../datasets/images/lenna.png')

#conner 
corner = cv2.cornerHarris(lenna,blockSize = 2,ksize =3 , k = 0.04)
lenna_copy = lenna.copy()
lenna_copy[corner > 0.01 * corner.max()] = [255]

lenna_copy = corner < corner.max() / 2 
plt.imshow(lenna_copy,cmap = 'gray')
```
