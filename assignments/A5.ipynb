### 1. Download the data from the link below using web scrapping: https://en.wikipedia.org/wiki/Mount_Everest

```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Mount_Everest'

headers = {
    'User-Agent': 'Custom-Script/1.9 (shivam@gmail.com)',
    'Accept-Language': 'en-US, en;q=0.5'
}

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
sentences = soup.text.split('\n')
sentences = [sentence for sentence in sentences if len(sentence.split(' ')) > 3]
text = '\n'.join(sentences)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Find the count of Total “Mount” in the text.

tokens = word_tokenize(text)

stop_words = stopwords.words('english')
words = [token.lower() for token in tokens if len(token) <= 20 and token not in stop_words and token not in string.punctuation]

print('Count of mount word:', words.count('mount'))
```

```python
# Print the title of this page.

print('Title of the page:', soup.find('title').text)
```

```python
# Save the extracted plain text contents in file mount.txt in upper case.

with open('mount.text', 'a', encoding='utf-8') as file:
    file.write(text.upper())
```

### 2. Read the following file

```python
from pypdf import PdfReader

reader = PdfReader('introduction-to-natural-language-processing.pdf')
```

```python
# Print only page number 12 and 13 from it.

page_12 = reader.pages[11].extract_text()
page_13 = reader.pages[12].extract_text()

text = 'Page 12:\n' + page_12 + '\n\nPage 13:\n' + page_13

print(text)
```

```python
# Save this contents in a file nlp.txt

with open('nlp.text', 'a', encoding='utf-8') as file:
    file.write(text)
```

### 3. Read the file from the link given below:
### https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx

```python
from spire.doc import Document

document = Document()
document.LoadFromFile('Machine-Learning-and-Natural-Language-Processing.docx')
text = document.GetText()
print('Document Text:\n', text)
```

```python
from nltk import sent_tokenize

# Read and display first 5 paragraphs from it.

sentences = sent_tokenize(text)
sentences = sentences[1:-1]

print('First 5 paragraphs:')

for i in range(len(sentences)):
    print(f'{i + 1}. {sentences[i]}')
```
