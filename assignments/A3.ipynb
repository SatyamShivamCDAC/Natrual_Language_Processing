```python
from nltk.corpus import wordnet
```

```python
# Read a word from user and print all meanings of it.

word = input('Enter a word: ')

synonyms = wordnet.synsets(word)

meanings = [f'{synonym.name()} -> {synonym.definition()}' for synonym in synonyms]

print(f'Meanings of {word}:\n', '\n'.join(meanings))
```

```python
# Read a word from user and print all noun meanings of it.

word = input('Enter a word: ')

synonyms = wordnet.synsets(word, pos='n')

meanings = [f'{synonym.name()} -> {synonym.definition()}' for synonym in synonyms]

print(f'Noun Meanings of {word}:\n', '\n'.join(meanings))
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser

# 1. Apply chuncking and show the chunks of type <VBD><DT>

pos_tags = pos_tag(word_tokenize(maharashtra))

rule = 'VBDT : {<VBD><DT>}'

parser = RegexpParser(rule)

tree = parser.parse(pos_tags)

tree.draw()
```

```python
import spacy
from spacy import displacy

# Find all the named entities present in article

spacy = spacy.load('en_core_web_sm')

text = spacy(maharashtra)

print('Named Entities:\n', text.ents)
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer
import string

# 3. Access all the verbs and apply the porter stemmer and lancaster stemmer

verbs = [word[0] for word in pos_tags if word[1].startswith('V') and word[0] not in string.punctuation]

stemmer = PorterStemmer()
stemmed_verbs = [stemmer.stem(verb.lower()) for verb in verbs]

print('Verbs after stemming (Porter):\n', stemmed_verbs)

stemmer = LancasterStemmer()
stemmed_verbs = [stemmer.stem(verb.lower()) for verb in verbs]

print('\nVerbs after stemming (Lancaster):\n', stemmed_verbs)
```

```python
from nltk.stem import WordNetLemmatizer

# 4. Access all the verbs in past tense and lemmatize them. Show only unique verbs.

past_verbs = {word[0] for word in pos_tags if word[1].startswith('VBD') and word[0] not in string.punctuation}

lemmatizer = WordNetLemmatizer()

lammatized_verbs = [lemmatizer.lemmatize(verb, pos='v') for verb in past_verbs]

print('Lemmatized Unique Past Verbs:', lammatized_verbs)
```

```python
from nltk.corpus import stopwords

# 5. Read only the adjectives from the text and find their synonyms.

stop_words = stopwords.words('english')

tokens = word_tokenize(maharashtra)
valid_words = {token.lower() for token in tokens if token.isalpha() and token not in stop_words}
pos_tags = pos_tag(valid_words)
adjectives = [word[0] for word in pos_tags if word[1].startswith('J')]

synonyms = [f'{adjective} -> {[synonym.name() for synonym in wordnet.synsets(adjective)]}' for adjective in adjectives]

print('Synonyms of Adjectives:\n', '\n'.join(synonyms))
```

```python
# 6. Read all the verbs and print their antonyms.

verbs = [word[0] for word in pos_tags if word[1].startswith('V') and word[0] not in string.punctuation]


verb_antonyms = dict()
for verb in verbs:
    antonyms = []
    for synonym in wordnet.synsets(verb):
        for lemma in synonym.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
                verb_antonyms[verb] = antonyms

verb_antonyms
```
