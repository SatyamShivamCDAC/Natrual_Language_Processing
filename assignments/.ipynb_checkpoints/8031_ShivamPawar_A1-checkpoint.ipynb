```python
from nltk.tokenize import word_tokenize, sent_tokenize
import string
```

```python
sentence = 'They told that their ages are 23, 25 and 29 respectively.'
tokens = word_tokenize(sentence)
```

### Q1

```python
# 1. Find the average of ages.

ages = [int(token) for token in tokens if token.isdigit()]
average = sum(ages) / len(ages)

print(f'Average of Ages: {average: .2f}')
```

```python
# 2. Count total number of words starts with 't' / 'T'

words = [token for token in tokens if token.lower().startswith('t')]

print('Words starts with t / T:', words)
print('Count of Words starts with t / T:', len(words))
```

```python
# 3. Count total punctuation symbols.

punctuations = [token for token in tokens if token in string.punctuation]

print('Punctuations:', punctuations)
print('Count of Punctuations:', len(punctuations))
```

```python
# 4. Print the words in capital letter which ends with 'd'

words = [token.upper() for token in tokens if token.lower().endswith('d')]

print('Words starts with d:', words)
```

```python
# 5. Merge all the words and print (no punctuations)

words = [token for token in tokens if token not in string.punctuation]
sentence = ' '.join(words)

print('Sentence without punctuation:', sentence)
```

### Q2 Open the wikipedia page of Indian flag and perform following analysis

```python
indian_flag = '''The national flag of India, colloquially called the Tiraṅgā (tricolour), is a horizontal rectangular tricolour flag, the colours being of India saffron, white and India green; with the Ashoka Chakra, a 24-spoke wheel, in navy blue at its centre.[1][2] It was adopted in its present form during a meeting of the Constituent Assembly held on 22 July 1947, and it became the official flag of the Union of India on 15 August 1947. The flag was subsequently retained as that of the Republic of India. In India, the term "tricolour" almost always refers to the Indian national flag. The current Indian flag was designed by Pingali Venkayya based on the Swaraj flag, a flag of the Indian National Congress adopted by Mahatma Gandhi after making significant modifications to the design proposed by Pingali Venkayya.[3] This flag included the charkha which was replaced with the chakra in 1947 by Tyabji.[4] Before the amendment of the flag code in 2021, the flag was by law only to be made of khadi; a special type of hand-spun cloth or silk, made popular by Mahatma Gandhi. The manufacturing process and specifications for the flag are laid out by the Bureau of Indian Standards. The right to manufacture the flag is held by the Khadi Development and Village Industries Commission, which allocates it to regional groups. As of 2023, there are four units in India that are licensed to manufacture the flag. Usage of the flag is governed by the Flag Code of India and other laws relating to the national emblems. The original code prohibited use of the flag by private citizens except on national days such as the Independence day and the Republic Day. In 2002, on hearing an appeal from a private citizen, Naveen Jindal, the Supreme Court of India directed the Government of India to amend the code to allow flag usage by private citizens. Subsequently, the Union Cabinet of India amended the code to allow limited usage. The code was amended once more in 2005 to allow some additional use including adaptations on certain forms of clothing. The flag code also governs the protocol of flying the flag and its use in conjunction with other national and non-national flags.'''
tokens = word_tokenize(indian_flag)
```

```python
# Count total number of Words

words = [token for token in tokens if not token.isdigit() and token not in string.punctuation]
print('Total number of words:', len(words))
```

```python
# Count percentage of punctuation marks

punctuations = [token for token in tokens if token in string.punctuation]

print('Count of Punctuations:', len(punctuations))
print('Percentage of Punctuations:', (len(punctuations) / len(tokens)) * 100)
```

```python
# Print total number of lines

sentences = sent_tokenize(indian_flag)

print('Total number of sentences:', len(sentences))
```

```python
# How many words are in title case?

words = [token for token in tokens if token[0].isupper()]

print('Words in Title case:', words)
print('\nCount of Words in Title case:', len(words))
```

```python
# Remove the numbers and print the text

words = [token for token in tokens if not token.isdigit()]
text = ' '.join(words)

print('Text without numbers:', text)
```
