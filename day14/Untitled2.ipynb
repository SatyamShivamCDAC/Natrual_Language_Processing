```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread('../datasets/images/dog.jpg')
plt.imshow(image)
```

```python
rows, cols = image.shape[:2]

rows, cols
```

```python
gridx, gridy = np.meshgrid(np.arange(cols), np.arange(rows))

gridy
```

```python
# define the mapping rule

mapx, mapy = gridx, gridy + 15 * np.sin(gridx / 20)

mapx = mapx.astype(np.float32)
mapy = mapy.astype(np.float32)
```

```python
plt.plot(np.arange(1, 40, 0.1), np.sin(np.arange(1, 40, 0.1)))
```

```python
mapped_image = cv2.remap(image, mapx, mapy, interpolation=cv2.INTER_LINEAR)
```

```python
mapped_image = cv2.cvtColor(mapped_image, cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(mapped_image)
```
