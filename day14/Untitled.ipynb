```python
import torch
import cv2
import numpy as np
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
cap = cv2.VideoCapture('../datasets/images/robots.mp4')

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print('End of video stream.')
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = model(frame_rgb)
    results.render()

    final_frame = cv2.cvtColor(results.ims[0], cv2.COLOR_RGB2BGR)
    cv2.imshow('YOLO v5 Object Detection', final_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
