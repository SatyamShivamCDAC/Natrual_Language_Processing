```python
from pypdf import PdfReader
from spire.doc import Document,FixedLayoutDocument
from nltk.tokenize import word_tokenize
```

<h1>Reading from PDF</h1>

```python
reader = PdfReader('../datasets/course.pdf')
```

```python
page1 = reader.pages[0] # first page of pdf 
```

```python
print(page1.extract_text()) # getting text written in first page of pdf 
```

```python
pgno = 0
for page in reader.pages:
    pgno += 1
    print('Page: ', pgno)
    print(page.extract_text())
```

```python
#copy the data from cousre.pdf and store in course.txt

with open('course.txt', 'a',encoding='utf8') as f:
    for page in reader.pages:
        f.write(page.extract_text())
```
