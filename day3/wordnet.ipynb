```python
#Extract specific POS from word

verbs = {x.lemmas()[0].name() for x in wordnet.synsets('sound', pos = 'v')}

noun = {x.lemmas()[0].name() for x in wordnet.synsets('sound', pos = 'n')}

adjective = {x.lemmas()[0].name() for x in wordnet.synsets('sound', pos = 'a')}


print(f'verbs -> {verbs}')
print(f'noun -> {noun}')
print(f'adjective -> {adjective}')
```

<h1>Antonyms</h1>

```python
#retrive the antonyms 
antonyms = set()
for word in wordnet.synsets('leopard'):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())
antonyms
```

<h1>Hypernyms</h1>

```python
#Extract the hypernyms 
word = 'soil'
sysnets = wordnet.synsets(word)

for word in sysnets:
    hypernyms = word.hypernyms()
    for h in hypernyms:
        print(word,'hypernym of',h.name())
```

<h1>Meronyms</h1>

```python
word = 'car'
sysnets = wordnet.synsets(word)

for x in sysnets:
    meronyms = x.part_meronyms()
    for m in meronyms:
        print(word, 'meronym of', m.name())
```

<h1>Holonymn</h1>

```python
word = 'keyboard'
sysnets = wordnet.synsets(word)

for x in sysnets:
    holonym = x.part_holonyms()
    for h in holonym:
        print(word,'holonym of', h.name())
```
