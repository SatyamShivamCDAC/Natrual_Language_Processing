```python
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk
```

```python
sent1 = 'This device is used to jam the signal'
sent2 = 'There is aheavy traffic jam here. We stuck at same place'
```

```python
a1 = lesk(word_tokenize(sent1), 'jam') #find the intended meaning of word

a1.definition()
```

```python
a2 = lesk(word_tokenize(sent2),'jam')

a2.definition()
```

```python
sent3 = 'It was a good catch by the player.'
sent4 = 'You have to catch that opportunity'

a3 = lesk(word_tokenize(sent3),'catch')
a4 = lesk(word_tokenize(sent4),'catch')

a3.definition(),a4.definition()
```

```python
sent5 = 'She like flowers'
sent6 = 'The association is like bad combination'

a5= lesk(word_tokenize(sent5),'like')
a6= lesk(word_tokenize(sent6),'like')
a5.definition(),a6.definition()
```
