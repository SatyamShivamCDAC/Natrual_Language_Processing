```python
import numpy as np

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import SimpleRNN,Dense,Input
from tensorflow.keras.models import Sequential
```

```python

alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
int_to_char = dict(list(enumerate(alphabets)))
int_to_char[6]
```

```python
char_to_int = {key :value for value,key in int_to_char.items()}
char_to_int
```

<h1>Preapare the datasets</h1>

```python
seq_len = 3

x = []
y = []

for i in range(len(alphabets) - seq_len):
    seq_in = alphabets[i:i+seq_len]
    seq_out = alphabets[i+seq_len]
    print(seq_in,'-->',seq_out)
    x.append([char_to_int[char] for char in seq_in]) 
    y += [char_to_int[seq_out]]
```

```python
len(x)
```

```python
x = np.reshape(x,(23,3,1))


x_norm  = x / 26
x_norm.shape
```

```python
y_new = to_categorical(y)
y_new.shape
```

<h1>Build the model</h1>

```python
model = Sequential()

model.add(Input((3,1)))
model.add(Dense(23,activation = 'relu'))
model.add(SimpleRNN(32))
model.add(Dense(26,activation = 'softmax'))

model.summary()
```

```python
model.compile(loss ='categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit(x_norm,y_new,epochs=500,batch_size = 1)
```

```python
model.evaluate(x_norm,y_new,batch_size=1)
```

```python
input_sequence = 'KLM'
abc = [char for char in input_sequence]
print(abc)
abc = [char_to_int[char] for char in input_sequence]
print(abc)
abc = np.reshape(abc,(1,3,1))/26
print(abc)
int_to_char[model.predict(abc).argmax(axis = 1)[0]]
```
