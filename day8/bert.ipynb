```python
import pandas as pd
import numpy as np

import tensorflow as tf
import ktrain
from ktrain import text

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('SMSSpamCollection',sep = '\t',names=['labels','text'])

df
```

```python
data_train,data_test = train_test_split(df,test_size=0.2,random_state = 0)
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
(X_train,y_train),(X_test,y_test),preproc = text.texts_from_df(
    train_df = data_train,
    text_column = 'text',
    label_columns = 'labels',
    val_df = data_test,
    preprocess_mode = 'bert'
)
```

```python
model = text.text_classifier(name = 'bert',train_data = (X_train,y_train),preproc= preproc)
```

```python
type(model)
```

```python
#learner object

learner = ktrain.get_learner(model = model,train_data = (X_train, y_train),
                            val_data = (X_test,y_test),batch_size = 6)
```

```python
learner
```

```python
learner.lr_find(show_plot=True,max_epochs = 2)
```

```python
predictor = ktrain.get_predictor(learner.model,preproc = preproc)
```

```python
sent = 'You won bonus prize of 100000. Call us to claim.'
```

```python
predictor.predict(sent)
```
