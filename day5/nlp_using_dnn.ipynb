```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
lines=[
    'It was a nice rainy day,',
    'The things were so beautiful in his point.',
    'When your focus is clear, you win.',
    'Many many happy returns of the day'
]
```

```python
lines
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines) #creates bag of word

tokenizer.word_docs 
```

```python
tokenizer.index_word
```

```python
tokenizer.word_index
```

```python
tokenizer.texts_to_matrix(lines) 
```

```python
tokenizer.word_index.keys() #tokens location
```

```python
seq = tokenizer.texts_to_sequences(lines) # encoding both index and posistion
seq
```

```python
padded = pad_sequences(seq,maxlen = 10,padding='pre') #padding wull add the remaing words of bag in each aar 
padded
```

```python
m = 0 
for sent in seq:
    if len(sent) > m:
        m = len(sent)
```

```python
m
```

```python
padded = pad_sequences(seq,maxlen=m,padding='post')
padded
```
