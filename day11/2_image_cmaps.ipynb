```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../datasets/images/dog.jpg',0)
```

```python
plt.figure(figsize = (18,12))
plt.subplot(1,2,1)
plt.imshow(img, cmap = 'Greys')

plt.subplot(1,2,2)
plt.imshow(img, cmap = 'Blues')
```

```python
orig_image = plt.colormaps['Greys'].reversed()
orig_image
```

```python
plt.figure(figsize = (16,16))
plt.subplot(3,3,1)
plt.title('Gray')
plt.imshow(img, cmap =  plt.colormaps['Greys'].reversed())

plt.subplot(3,3,2)
plt.title('Blue')
plt.imshow(img, cmap =  plt.colormaps['Blues'].reversed())

plt.subplot(3,3,3)
plt.title('Red')
plt.imshow(img, cmap =  plt.colormaps['Reds'].reversed())


plt.subplot(3,3,4)
plt.title('Purple')
plt.imshow(img, cmap =  plt.colormaps['Purples'].reversed())

plt.subplot(3,3,5)
plt.title('Green')
plt.imshow(img, cmap =  plt.colormaps['Greens'].reversed())



plt.subplot(3,3,6)
plt.title('Orange')
plt.imshow(img, cmap =  plt.colormaps['Oranges'].reversed())



plt.subplot(3,3,7)
plt.title('PuBu')
plt.imshow(img, cmap =  plt.colormaps['PuBu'].reversed())

plt.subplot(3,3,8)
plt.title('YlOrBr')
plt.imshow(img, cmap =  plt.colormaps['YlOrBr'].reversed())

plt.subplot(3,3,9)
plt.title('BuGn')
plt.imshow(img, cmap =  plt.colormaps['BuGn'].reversed())


```
