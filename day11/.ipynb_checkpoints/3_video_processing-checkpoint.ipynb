<h1>Frame at Specific timestamp -> trigger frame</h1>

```python
time = '00:00:45'
```

```python
# frame number at time stamp = sec_intv *fps

hrs,mins,secs = time.split(':')
sec_intv = int(hrs) * 3600 + int(mins) * 60 + int(secs)
trigger_frame = sec_intv * fps 
trigger_frame
```

```python
#stroing trigger frame
vid = cv2.VideoCapture('../datasets/images/nature.mp4')
img = 0
while True:
    ret , frame = vid.read()
    if ret == False:
        break
    if img == trigger_frame:
        cv2.imwrite(f'images/trigger_frames/frame{img}.jpg',frame)
    img += 1
    
    # cv2.imshow('my vid', frame)
    # if cv2.waitKey(10) == ord('q'): #waitkey will wait for given milli seconds before loading next frame 
    #     break
cv2.destroyAllWindows()
```
