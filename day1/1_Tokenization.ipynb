<h1 style = 'color : orange'>Word Punct Tokenizer</h1>

```python
tk = WordPunctTokenizer() #initializer word punt tokenizer
sent = 'They told that their age are 26 31, 29 and 29 respectively'
tk.tokenize(sent)
```

<h1 style = 'color : orange'>Space Tokenizer</h1>

```python
tk = SpaceTokenizer() #only separate for blank spaces
sent = 'They told that their age are 26 31, 29 and 29 respectively'
tk.tokenize(sent)
```

<h1 style = 'color : orange'>Tab Tokenizer</h1>

```python
tk = TabTokenizer() #only separate for tab spaces
sent = 'They told that their age are 26 31, 29 and 29 respectively\tthis message'
tk.tokenize(sent)
```

<h1 style = 'color : orange'>Line Tokenizer</h1>

```python
sentence = """
Hello! 
How are You?
I am Fine
"""

tk = LineTokenizer()
tk.tokenize(sentence)
```

<h1 style = 'color : orange'>Whitespace Tokenizer</h1>

```python
sentence = """
Hello! 
How are You? Where are you from?
I am Fine
"""

tk = WhitespaceTokenizer()
tk.tokenize(sentence)
```

<h1 style = 'color : orange'>Multi-word Express Tokenizer</h1>

```python
nsent = 'Chhrapati Shivaji Maharaj was founder of Maratha Empire'
tk = MWETokenizer(separator= ' ')
tk.add_mwe(('Chhrapati', 'Shivaji')) # combine two tokens into one
tk.tokenize(word_tokenize(nsent)) # combining tokens from list of tokens 

```

<h1 style = 'color : orange'>Tweet Tokenizer</h1>

```python
nsent = 'Hello Friends!:) How are you? Welcome to Python <3Programming'
tk = TweetTokenizer()
tk.tokenize(nsent) # combining tokens from list of tokens 
```
