```python
nltk.download('punkt') # tokenization
nltk.download('punkt_tab') #tokenization
nltk.download('averaged_perceptron_tagger') # POS tagging
nltk.download('averaged_perceptron_tagger_eng') # POS tagging

nltk.download('wordnet') # worldnet vocabulary
nltk.download('stopwords') # stopwords vocabulary
nltk.download('omw-1.4') # stemming and lemmatization
nltk.download('indian') # Indian language POS tagging
nltk.download('maxent_ne_chunker') # NER Chunking
nltk.download('maxent_ne_chunker_tab') # NER Chunking
```

```python
sent = 'They told that their age are 26 31 and 29 respectively'
# write a program to find average of numbers in sentence
import re

nums = re.findall(r'\d.',sent)
summ = 0
count = 0
for i in nums:
    summ += int(i)
    count += 1
print('average is', summ/count)
```
