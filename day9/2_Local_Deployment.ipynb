```python
import joblib
from tkinter import *
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

sword = stopwords.words('english')
ps = PorterStemmer()


def clean_text_stem(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [ps.stem(token) for token in tokens2 if token.lower() not in sword]
    return tokens3
```

```python
spam_detector = joblib.load('spam_detector.model')
preprocessor = joblib.load('spam_detector_preprocessor.model')
# clear_text = joblib.load('clean_text.model')
```

```python
top = Tk()
top.title('Spam Message Detection System')
top.geometry('800x500')


def check():
    message = t.get('1.0',END)
    message = preprocessor.transform([message]) #preprocessor for model
    pred = spam_detector.predict(message) #using model to predict 
    result.config(text = f'message is {pred[0]}') #displaying result on screen
    print(pred)  

l1 = Label(top,text = 'Enter your message here.') 
t = Text(top)
l1.place(x = 100,y = 10)
t.place(x=100,y = 40)
b = Button(top,text = 'Check Mesage', command = check).place(x= 100,y =450)

result = Label(top,text = '')
result.place(x = 100,y = 480)
top.mainloop()
```
