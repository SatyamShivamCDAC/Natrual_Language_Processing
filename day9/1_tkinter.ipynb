```python
top = Tk()

def find():  #function that will be called when button is clicked
    num = int(e.get())  #number from ebtry box
    result = Label(top,text = f'Sqaure of the {num} is '+ str(num**2))
    result.pack()
    
l = Label(text = 'Enter the number ') #label object
e = Entry() #Entry object to tkae user input
b = Button(top,text= 'Click here',command = find) #button to perform action, command will hit function 
top.geometry('800x500')

l.pack()
e.pack()
b.pack()
top.mainloop()
```

```python
top = Tk()

def find():  #function that will be called when button is clicked
    num = int(e.get())  #number from ebtry box
    result.config(text = f'Square is {num**2}') #re-enter text for result
    
l = Label(text = 'Enter the number ') #label object
e = Entry() #Entry object to tkae user input
b = Button(top,text= 'Click here',command = find) #button to perform action, command will hit function 
result = Label(top,text = '')  #result label to show sqare number 
top.geometry('400x200')

l.place(x = 50,y = 50)
e.place(y = 80,x = 50)
b.place(y =  100,x = 50)
result.place(y = 130,x = 50)
top.mainloop()
```

```python
top = Tk()

def addi():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text = f'Addition is {num1 + num2}')

def subs():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text = f'Substraction is {num1 - num2}')
def muls():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text = f'Multiplication is {num1 * num2}')
def divs():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text = f'Division is {num1 / num2}')



l1 = Label(top,text = 'Number 1')
e1 = Entry()
l2 = Label(top,text = 'Number 2')
e2 = Entry()

add = Button(top,text = 'Add',command = addi)
sub = Button(top,text = 'Sub',command = subs)
div = Button(top,text = 'Div',command = divs)
mul = Button(top,text = 'Mul',command = muls)

result = Label(top,text = '')

top.geometry('700x350')

l1.place(x = 200, y = 100)
e1.place(x = 300, y = 100)
l2.place(x = 200, y = 130)
e2.place(x = 300, y = 130)

add.place(x = 150,y = 180)
sub.place(x = 250,y = 180)
div.place(x = 350,y = 180)
mul.place(x = 450,y = 180)

result.place(x = 300,y = 210)
top.mainloop()
```
