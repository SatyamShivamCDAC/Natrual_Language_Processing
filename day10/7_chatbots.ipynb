```python
import nltk 
import random
```

```python
#define the chatbot responses and intents 
reponses = {
   'greeting' : ['Hi!','Hello!','Namaste!','Hi there!'],
    'goodbye' : ['Good Bye!','See you later!','Take care','Bye!'],
    'thanks' : ['Your\'re welcome', 'Glad to help', 'Anytime','Happy to assist'],
    'default' : ['Sorry I didnt understand', 'Can yo tell me again?','I am not sure.']
}
```

```python
#simple set of questions and reponses
patterns = {
    'hi':'greeting',
    'hello':'greeting',
    'hey' : 'greeting',
    'bye' : 'goodbye',
    'thanks' : 'thanks',
    'thank you' : 'thanks'
}
```

```python
# match user input 
def respond(user_input):
    for word in nltk.tokenize.word_tokenize(user_input.lower()):
        if word in patterns:
            intent = patterns[word]
            return random.choice(reponses[intent])
    return random.choice(reponses['default'])
```

```python
respond('Hi')
```

```python
respond('Ky are ')
```

```python
def chat():
    print('ChatBot: How can I help you today? (Type exit to quit)')
    while True:
        user_input = input('You: ')
        if user_input == 'exit':
            print('ChatBot: Good Bye!')
            break
        response = respond(user_input)
        print('ChatBot:', response)
```

```python
chat()
```
