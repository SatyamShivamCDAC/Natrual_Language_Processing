```python
import pyttsx3
import speech_recognition as sr
```

<h1>TTS</h1>

```python
engine = pyttsx3.init()
```

```python
#set the properties 
engine.setProperty('rate',150)
engine.setProperty('volume',0.9)
```

```python
text = 'Welcome to world of Deep learning using Python Progamming'
engine.say(text)
engine.runAndWait()
```

<h1>ASR</h1>

```python
r = sr.Recognizer()
```

```python
with sr.Microphone() as source:
    r.adjust_for_ambient_noise(source)
    print('Speak : ')
    audio = r.listen(source)
    print('You said:' , r.recognize_google(audio))
```
