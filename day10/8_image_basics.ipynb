```python
imgz = np.zeros([500,500],dtype = np.uint8)
plt.imshow(imgz,cmap='gray')
```

```python
imgz[250,:] = 255
plt.imshow(imgz,cmap = 'gray')
```

```python
imgz[:,250] = 255
plt.imshow(imgz,cmap = 'gray')
```

```python
imgz[:100,:100] = 200
plt.imshow(imgz,cmap = 'gray')
```

```python
image_sam = np.zeros([10000],dtype = np.uint8)
for i in range(10000):
    image_sam[i] = np.random.randint(0,255)
```

```python
image_sam = image_sam.reshape(100,100)
plt.imshow(image_sam,cmap = 'gray')
```
