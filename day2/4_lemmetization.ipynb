```python
from nltk.stem import WordNetLemmatizer
```

```python
word1 = 'consultant','consult','consultants','consulting','consultative'
word2 = 'management','manage','manager','managing','managed'
word3 = 'big','bigger','biggest'
word4 = 'go','gone','went','going'
word5 = 'association', 'associating', 'associated', 'associations'
```

```python
wnl = WordNetLemmatizer()
```

```python
for word in word1:
    print(wnl.lemmatize(word,pos = 'v'))
```

```python
for word in word3:
    print(wnl.lemmatize(word,pos = 'a'))
```

```python
print(wnl.lemmatize('is',pos = 'v'))
print(wnl.lemmatize('children',pos = 'n'))
print(wnl.lemmatize('worst',pos = 'a'))
```
