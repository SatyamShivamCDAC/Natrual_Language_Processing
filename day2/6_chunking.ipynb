```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag,RegexpParser
```

```python
grammer = 'NP : {<DT><JJ><NN>}'

sent = 'It was a beatifull game by him. An amazing play was witnessed'

chunker = RegexpParser(grammer) 

tags = pos_tag(word_tokenize(sent)) #extracting the tags

tags
```

```python
tree = chunker.parse(tags)
tree.draw()
```
