```python
from string import punctuation
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
#remove punchation from sentence

sent = 'Dr. friends! How are you? Welcome to Progamming.'

tokens = word_tokenize(sent)
out = [x for x in tokens if x not in punctuation]
' '.join(out)
```

<h1 style = 'color:orange'>Stopword Removal</h1>

```python
swords = stopwords.words('english') #printing stopwords for english language
swords
```

```python
# removing sstopwords from sentence

sent = 'Hello friends! How are you? Welcome to Progamming.'

swords = stopwords.words('english')

keywords = [x for x in word_tokenize(sent) 
            if (x.lower() not in swords) and (x not in punctuation)]
keywords
```
