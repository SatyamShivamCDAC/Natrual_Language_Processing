```python
from nltk.stem import PorterStemmer,LancasterStemmer
```

```python
word1 = 'consultant','consult','consultants','consulting','consultative'
word2 = 'management','manage','manager','managing','managed'
word3 = 'big','bigger','biggest'
word4 = 'go','gone','went','going'
word5 = 'association', 'associating', 'associated', 'associations'
```

<h1 style = 'color : orange'>Porter Stemmer</h1>

```python
#creating object
ps = PorterStemmer()
ps.stem(word2[3])
```

```python
for word in word1:
    print(ps.stem(word))
```

```python
for word in word4:
    print(ps.stem(word))
```

<h1 style = 'color : orange'>Lancaster Stemmer</h1>

```python
ls = LancasterStemmer()
```

```python
print(word3)
print()
for word in word3:
    print(ls.stem(word))
    print(ps.stem(word))
    print()
    
```
